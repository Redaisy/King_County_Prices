```python
#Preparation to plot regplot between sqft_living and price, which will visualize regression line and confidence interval
#separate dataframe into separate dataframes, one for each grade
#use loc to separate by grade = ?
grade_dfs = []
grades = list(df.grade.value_counts().index)
grades.sort()
for num in grades:
    name_of_df = 'df_'+str(num) 
    name_of_df = df.loc[df['grade'] == num]
    grade_dfs.append(name_of_df)
    
#grade_dfs is list which holds a dataframe for each grade
#Plotting regplot between sqft_living and price for each grade by iterating through grade_dfs

sns.set(style="whitegrid")

for i in range(0 , len(grade_dfs)):
    f, axs = plt.subplots(figsize = (8,6))
    sns.regplot(x='sqft_living', y="price", data=grade_dfs[i], scatter_kws={"color": "green", 'alpha': 0.3}, line_kws={"color": "red"})
    plt.title('Square Feet vs. Price for Houses of Grade ' +str(i+3), fontsize = 15)
    plt.show()
```
